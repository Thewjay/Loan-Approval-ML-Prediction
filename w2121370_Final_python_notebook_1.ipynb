**Reused from: Code Reuse Session 1 (Prompts 17 & 19)**

```python
# Build a function to detect outliers in loan_interest_rate variable using the IQR
def find_outliers_IQR(data):

  # Calculate the first quartile (25th percentile)
  q1=data.quantile(0.25)

  # Calculate the third quartile (75th percentile)
  q3=data.quantile(0.75)

  # Calculate the Inter-Quartile Range
  IQR=q3-q1

  # Filter the data to find data points outside the defined boundaries
  outliers = data[((data<(q1-1.5*IQR)) | (data>(q3+1.5*IQR)))]

  # Return the detected outliers
  return outliers

# Call the built IQR function to detect outliers in loan_interest_rate variable
outliers = find_outliers_IQR(data['loan_interest_rate'])

# Print the total number of detected outliers
print("number of outliers: "+ str(len(outliers)))

# Display the detected outliers
outliers
```

**Reused from: Code Reuse Session 1 (Prompts 12, 23, 24, 25, 26)**

```python
# Remove unnecessary variables
data = data.drop(columns=(['id', 'loan_income_ratio']))

# Map the categorical values 'Y' & 'N' to numerical format '0' & '1' for the model to process
data['payment_default_on_file'] = data['payment_default_on_file'].map({'N': 0 , 'Y': 1})

# Convert float variable 'age' into an integer data type as the decimal precision is not important
data = data.astype({"age":"int"})

# Import the preprocessing package from sklearn to encode ordinal columns into numeric labels
from sklearn import preprocessing

# Initialize the LabelEncoder
label_encoder = preprocessing.LabelEncoder()

# Apply the method to encode the object type variables
data['home_ownership'] = label_encoder.fit_transform(data['home_ownership'])
data['loan_intent'] = label_encoder.fit_transform(data['loan_intent'])
```

**Reused from: Code Reuse Session 1 (Prompts 12, 29, 30, 32, 33, 28, 27)**

```python
# Group the numeric features into one variable
numeric_cols = ['age', 'income', 'emplyment_length', 'loan_amount', 'loan_interest_rate', 'credit_history_length']

# Group the categorical features into one variable
categorical_cols = ['home_ownership', 'loan_intent', 'payment_default_on_file']

# Import the MinMaxScaler package
from sklearn.preprocessing import MinMaxScaler

# Initialize the MinMaxScaler
scaler = MinMaxScaler()

# Apply Minimum-Maximum Scaling to the numeric columns
data[numeric_cols] = scaler.fit_transform(data[numeric_cols])

# Reuse prompt 12 to remove regression target before saving the classification report
classification_data = data.drop(columns=['max_allowed_loan'])

# Save the prepared classification dataset in a CSV file
classification_data.to_csv(r'Classification_Loan_Data.csv', index=False)

# Isolate the clients who were approved for a loan
regression_data=data[data.loan_approval_status == 0]

# Save the prepared regression dataset in a CSV file
regression_data.to_csv(r'Regression_Loan_Data.csv', index=False)
```
